## Magnetic fields and magnetization (SI units)

In magnetism, we work with three closely related vector fields:

- **Magnetic field** $\mathbf{H}$ — the field produced by electric currents and magnetic sources  
  Units: amperes per meter (A m⁻¹)

- **Magnetization** $\mathbf{M}$ — the magnetic moment per unit volume carried by matter  
  Units: amperes per meter (A m⁻¹)

- **Magnetic induction** $\mathbf{B}$ — the total magnetic field that results from both $\mathbf{H}$ and the material response  
  Units: tesla (T)

In SI units these are related by

$
\mathbf{B} = \mu_0 \, (\mathbf{H} + \mathbf{M}),
$

where $\mu_0 = 4\pi \times 10^{-7}\ \mathrm{H\,m^{-1}}$ is the permeability of free space.  

This equation makes explicit that $\mathbf{B}$ includes both the externally applied field $\mathbf{H}$ and the field created by the magnetization of the material itself.

### Units

To understand the units, let's consider the magnetic field at the center of a circular current loop of radius $r$ carrying current $i$. The loop produces a magnetic field $\mathbf{H}$ at its center.

$
H = \frac{i}{2r} \mathrm{\;[Amperes/meter, A /m]}
$

The current loop has a magnetic moment ($m$) associated with it which is:

$
m=i*area \mathrm{\;[A m^2]}
$

where $area$ is the area of the loop. Given that current $i$ has units of A and $area$ has units of $m^2$, magnetic moment has units of $\mathrm{\;[A m^2]}$

We can consider a magnetic material (like a rock), to be a volume containing numerous tiny current loops (atomic magnetic moments). We quantify the material's magnetic strength as **Magnetization ($\mathbf{M}$)**, which is defined as the net magnetic dipole moment per unit volume ($V$):

$
M = \frac{m}{V} \mathrm{\;[A /m]}
$

Given that magnetic moment has units of $\mathrm{A m^2}$ and volume has units of $\mathrm{m^3}$, we end up with units for magnetization of $\mathrm{A/m}$. Note that the units of **Magnetic field** ($\mathbf{H}$) and **Magnetization** $\mathbf{M}$ are both A/m.

### Thinking about $\mathbf{B}$

How should we conceptualize the magnetic induction $\mathbf{B}$? While $\mathbf{H}$ represents the external "driving" influence and $\mathbf{M}$ represents the material's internal response, $\mathbf{B}$ represents the net physical density of magnetic flux lines passing through the medium. It is the vector sum of the drive and the response, scaled by the permeability of free space. Crucially, $\mathbf{B}$ is the quantity that interacts with the physical world: it exerts the Lorentz force on moving charges and induces voltages in wire loops. Consequently, when we use a magnetometer to measure the Earth's magnetic field or a rock's remanence, we are measuring $\mathbf{B}$. Because geomagnetic fields are weak, we typically quantify $\mathbf{B}$ in nanotesla (nT), where $1 \text{ nT} = 10^{-9} \text{ T}$ or microtesla ($\mu \text{T}$), where $1\;\mu \text{T} = 10^{-6} \text{ T}$.

## Magnetic Susceptibility

How strongly a material reacts to a magnetic field is one of its most fundamental physical properties. When a material is placed in a magnetic field $\mathbf{H}$, it develops an internal magnetization $\mathbf{M}$. For most geological materials in weak to moderate fields, this response is linear and directly proportional to the applied field.

We describe this relationship using a parameter called the **volume magnetic susceptibility**, denoted by the Greek letter $\kappa$ (kappa):

$$\mathbf{M} = \kappa \mathbf{H}$$

Because both $\mathbf{M}$ and $\mathbf{H}$ have units of Amperes per meter (A/m), volume susceptibility $\kappa$ is a **dimensionless quantity** in the SI system. It represents a volume-normalized efficiency ratio: how much magnetization is generated per unit of applied field.

### Physical Classification of Materials
The value and sign of $\kappa$ provide a window into the atomic-scale magnetic interactions inside a mineral:

1.  **Diamagnetism ($\kappa < 0$):** These materials have a small, negative susceptibility. The induced magnetization opposes the applied field. Common examples include quartz, calcite, and water.
2.  **Paramagnetism ($\kappa > 0$):** These materials have a small, positive susceptibility. Unpaired electron spins partially align with the field, slightly amplifying it. Common examples include iron-bearing silicates like olivine and pyroxene.
3.  **Ferromagnetism ($\kappa \gg 0$):** These materials have a very large, positive susceptibility. Strong quantum mechanical interactions between spins produce intense magnetization even in weak fields. Common examples include magnetite (a ferrimagnet) and hematite (a canted antiferromagnet), which are broadly referred to as ferromagnetic because they possess spontaneous magnetization.

### Geological Significance
In Earth materials, susceptibility is a powerful diagnostic tool because it links quick bulk magnetic measurements to mineralogy. Because ferromagnetic minerals (like magnetite) have susceptibility values orders of magnitude higher than paramagnetic or diamagnetic minerals, even trace amounts of magnetite can dominate the bulk susceptibility of a rock. Consequently, susceptibility is often used as a proxy for the concentration of magnetic minerals within a geological sample.

## Diamagnetism

Diamagnetism is a fundamental magnetic property of all matter. It arises from the orbital
motion of electrons around atomic nuclei. An orbiting electron constitutes a microscopic
current loop and therefore produces a magnetic moment. When an external magnetic field is
applied, the Lorentz force slightly modifies the electron’s orbital motion, changing the
strength of this current and inducing an additional magnetic moment.

The induced magnetic moment always opposes the applied magnetic field, producing a weak
negative magnetization. This opposition is a direct consequence of electromagnetic
induction and does not depend on the existence of permanent atomic magnetic moments.

### Characteristics of diamagnetism

Because diamagnetism arises from changes in electronic orbital motion rather than the
thermal alignment of magnetic dipoles, its susceptibility is nearly temperature independent
in most solids. Thermal agitation does not disrupt this induced orbital response in the way
it affects paramagnetic or ferromagnetic alignment.

Diamagnetism is present in every atom, but it is very weak. In materials that contain
unpaired electrons or magnetically ordered phases, stronger paramagnetic, ferromagnetic,
or ferrimagnetic contributions usually dominate the bulk magnetic behavior. Diamagnetic
behavior is therefore most apparent in substances composed primarily of closed-shell ions
or molecules.

### Geological significance of diamagnetism

Most major rock-forming minerals are diamagnetic. Quartz (SiO₂), calcite (CaCO₃), and
feldspars all have negative magnetic susceptibilities. As a result, clean sandstones,
limestones, and other rocks composed mainly of these minerals can exhibit net negative
magnetic susceptibility if they contain little iron-bearing material.

Typical diamagnetic susceptibilities of common minerals are small and negative, generally
on the order of \(10^{-6}\) to \(10^{-5}\) in SI units.

## Paramagnetism

Paramagnetism arises in materials that contain atoms with **unpaired electrons**. Unlike the filled shells of diamagnetic atoms, these unpaired electrons act like tiny atomic bar magnets (magnetic dipoles) with a permanent magnetic moment.

In the absence of an external magnetic field, these atomic dipoles point in random directions due to thermal agitation (heat energy). As a result, the material has **zero net magnetization**.

When an external field ($\mathbf{H}$) is applied, it exerts a torque on these dipoles, encouraging them to align with the field. This alignment creates a net positive magnetization ($\mathbf{M}$) that adds to the external field.

### Characteristics of Paramagnetism
* **Response:** The induced magnetization is positive and linear with the applied field (for weak fields).
* **Susceptibility ($\kappa$):** Positive, typically in the range of $10^{-5}$ to $10^{-3}$ (SI).
* **Temperature dependence:** Strong. Thermal energy ($kT$) constantly competes with magnetic
alignment. As temperature increases, thermal motion randomizes the orientations of atomic
magnetic moments, reducing the net magnetization induced by a field. For many paramagnetic
materials this behavior is approximately described by Curie's law,
$
\kappa \approx \frac{C}{T},
$
where $C$ is the Curie constant (dependent on the material) and $T$ is temperature in kelvin. More generally, weak
interactions between neighboring moments modify this to the Curie-Weiss law,
$
\kappa = \frac{C}{T - \theta},
$
where $theta$ (the Weiss parameter) reflects the strength and sign of these interactions.

### Geological Significance
Paramagnetism is the dominant magnetic behavior of iron-bearing silicate minerals. **Olivine** ($(Mg,Fe)_2SiO_4$), **pyroxene**, **amphibole**, and **biotite** are all paramagnetic.

Because the magnetic moment comes from the iron ($Fe^{2+}$ or $Fe^{3+}$), the strength of the susceptibility is often directly proportional to the iron content.
* **Forsterite ($Mg_2SiO_4$):** The iron-free endmember of olivine is actually **diamagnetic** ($\kappa \approx -1.2 \times 10^{-5}$).
* **Fayalite ($Fe_2SiO_4$):** The iron-rich endmember is strongly **paramagnetic** ($\kappa \approx 5.5 \times 10^{-3}$).
* **Typical Mantle Olivine:** Most natural olivine contains enough iron to be paramagnetic, with typical susceptibility values around $\kappa \approx 1.6 \times 10^{-3}$.

## Visualizing Linear Magnetic Response

To visualize the difference between diamagnetism and paramagnetism, we can simulate a standard rock magnetic experiment known as a **Hysteresis Loop**.

In a real laboratory measurement, we vary the applied magnetic field to trace a complete cycle:
1.  Ramp up to a maximum positive field (e.g., $+1$ Tesla).
2.  Ramp down to a maximum negative field (e.g., $-1$ Tesla).
3.  Return to zero.

For **diamagnetic** and **paramagnetic** materials, this process is reversible. There is no "magnetic memory" retained when the field goes to zero. Consequently, the "loop" collapses into a straight line passing through the origin.

The function below simulates this cycle. It takes the volume susceptibility ($\kappa$) as an input and calculates the resulting magnetization as a function of a varying field.

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_linear_hysteresis(chi_si, save_as=None):
    """
    Plots the hysteresis loop (M vs B) for a material with a linear magnetic response.

    Parameters:
    chi_si (float): Volume magnetic susceptibility (dimensionless, SI units).
    save_as (str, optional): Filename to save the plot (e.g. 'augite.png').
    """

    mu0 = 4 * np.pi * 1e-7  # Permeability of free space (H/m)

    # Define the Applied Field Loop (in Tesla)
    # Trace path: 0 -> 1T -> -1T -> 0
    B_up = np.linspace(0, 1, 50)
    B_down = np.linspace(1, -1, 100)
    B_return = np.linspace(-1, 0, 50)
    B_applied = np.concatenate([B_up, B_down, B_return])

    # Calculate Magnetization: M = chi * H
    # Note: Since B = mu0 * H (approx in air), H = B / mu0
    H_applied = B_applied / mu0
    M = chi_si * H_applied

    # Determine material type for the label
    if chi_si < 0:
        material_type = "Diamagnetic"
    else:
        material_type = "Paramagnetic"

    plt.figure(figsize=(8, 6))

    label_text = f"{material_type} specimen\n" + rf"$\kappa = {chi_si:.1e}$"
    plt.plot(B_applied, M, color='blue', linewidth=3, label=label_text)

    plt.axhline(0, color='black', linewidth=1, linestyle='-')
    plt.axvline(0, color='black', linewidth=1, linestyle='-')
    plt.grid(True, linestyle='--', alpha=0.5)

    plt.title(f"Hysteresis Loop: {material_type} Response", fontsize=14)
    plt.xlabel(r"B ($\mu_0 H$), Applied Field (T)", fontsize=12)
    plt.ylabel("M, Magnetization (A/m)", fontsize=12)
    plt.legend(fontsize=12)

    if save_as:
        if not save_as.endswith(('.png', '.jpg', '.pdf', '.svg')):
            save_as += ".png"

        plt.savefig(save_as, dpi=300, bbox_inches='tight')

    plt.show()
```

### Example 1: Quartz (Diamagnetism)

First, let's look at **Quartz**, a common diamagnetic mineral. It has a small, negative susceptibility ($\kappa \approx -1.5 \times 10^{-5}$).

Run the code below to see its magnetic response. Notice that the slope is negative, meaning the material slightly opposes the applied field. Additionally, while we are calling it a "hysteresis loop," in reality it collapses to a single straight line because diamagnetic and paramagnetic materials have no magnetic memory (no remanence or coercivity).

```python
# Define susceptibility for Quartz (Diamagnetic)
chi_quartz = -1.5e-5

# Plot the hysteresis "loop"
plot_linear_hysteresis(chi_quartz, save_as="quartz_hysteresis.png")
```
