## Background

Paleomagnetic directions are described by **declination** ($D$) and **inclination** ($I$). Under the **geocentric axial dipole (GAD)** hypothesis, the time-averaged geomagnetic field is that of a dipole aligned with Earth's spin axis, and the field inclination is related to geographic latitude by:

$$\tan(I) = 2\tan(\lambda)$$

This relationship is the foundation of paleolatitude determination: measure the inclination of ancient rocks, invert the dipole formula, and you know the latitude at which the rock formed.

**Thermoremanent magnetization (TRM)** in lavas faithfully records the ambient field direction. However, **depositional remanent magnetization (DRM)** in sediments is subject to **inclination shallowing** — a systematic bias where the recorded inclination is shallower than the true field inclination. This occurs because elongate or platy magnetic grains tend to lie flat on the bedding plane during deposition and compaction.

The flattening formula (King, 1955) describes this bias:

$$\tan(I_{\text{DRM}}) = f \cdot \tan(I_{\text{GAD}})$$

where $f$ is the **flattening factor** ($f = 1$: no shallowing; $f < 1$: inclination is shallowed). Typical values in natural sediments range from about 0.4 to 0.8.

### Useful `pmagpy` functions

You will need the following functions from `pmagpy.ipmag`:

- `ipmag.inc_from_lat(lat)` — returns the GAD inclination for a given latitude (in degrees)
- `ipmag.lat_from_inc(inc)` — returns the paleolatitude for a given inclination (in degrees)
- `ipmag.squish(inc_list, f)` — applies the flattening formula to a list of inclinations with flattening factor `f`; returns a list of shallowed inclinations

Try them out:

```python
# Example: GAD inclination at 45° latitude
inc_45 = ipmag.inc_from_lat(45)
print(f"GAD inclination at 45° latitude: {inc_45:.1f}°")

# Example: Apply flattening with f=0.6
inc_drm = ipmag.squish([inc_45], 0.6)
print(f"DRM inclination (f=0.6): {inc_drm[0]:.1f}°")

# Example: Invert to get paleolatitude
plat = ipmag.lat_from_inc(inc_drm[0])
print(f"Paleolatitude from DRM inclination: {plat:.1f}° (true: 45.0°)")
```
